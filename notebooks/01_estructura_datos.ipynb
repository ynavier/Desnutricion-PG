### Cargar los datasets desde data/raw

```python
data2023 = pd.read_excel("../data/raw/2023_113_Reporte.xlsx")
data2024 = pd.read_excel("../data/raw/2024_113_Reporte.xlsx")
data2025 = pd.read_excel("../data/raw/2025_113_Reporte.xlsx")
```

### Agrupar datasets en un diccionario

```python
datasets = {
    "2023": data2023,
    "2024": data2024,
    "2025": data2025
}
```

### Revisar dimensiones de los datasets

```python
for year, df in datasets.items():
    print(f"\n===== DIMENSIONES DATASET {year} =====")
    print("Filas y columnas:", df.shape)
```

### Revisar tipos de datos

```python
for year, df in datasets.items():
    print(f"\n===== TIPOS DE DATOS {year} =====")
    print(df.dtypes)
```

### Ver columnas de cada dataset

```python
for year, df in datasets.items():
    print(f"\n===== COLUMNAS DATASET {year} =====")
    print(list(df.columns))
```

### Revisar valores nulos

```python
for year, df in datasets.items():
    print(f"\n===== NULOS DATASET {year} =====")
    print(df.isnull().sum().sort_values(ascending=False))
```

### Ver solo columnas con nulos

```python
for year, df in datasets.items():
    print(f"\n===== COLUMNAS CON NULOS {year} =====")
    
    nulls = df.isnull().sum()
    nulls = nulls[nulls > 0].sort_values(ascending=False)
    
    print(nulls)
```

### Porcentaje de datos faltantes

```python
for year, df in datasets.items():
    print(f"\n===== PORCENTAJE DE NULOS {year} =====")
    
    percent = (df.isnull().sum() / len(df)) * 100
    print(percent.sort_values(ascending=False))
```

### Revisar duplicados

```python
for year, df in datasets.items():
    print(f"\n===== DUPLICADOS DATASET {year} =====")
    print("Registros duplicados:", df.duplicated().sum())
```

### Ver primeras filas

```python
for year, df in datasets.items():
    print(f"\n===== PRIMERAS FILAS DATASET {year} =====")
    display(df.head())
```

### Comparar si las columnas son iguales entre datasets

```python
print("¿Columnas iguales entre 2023 y 2024?")
print(set(data2023.columns) == set(data2024.columns))

print("\n¿Columnas iguales entre 2023 y 2025?")
print(set(data2023.columns) == set(data2025.columns))
```
